```python
import re
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.fftpack

filename = '/home/hador/workspace/pleasedKit/records/flame-labjack-gain10'
#filename = '/home/hador/workspace/pleasedKit/records/flame-2014-08-26-2'
OFFSET = 0.05
```

```python
data1=[]
data2=[]
timeCounter=0

with open(filename,'r') as f:
    for line in f:
        # if (re.match('>', line) and not re.match('>,', line) and not re.search(',\n$', line)):
        if (re.match('>', line)):
            s = line.translate(None, '>\n').split(',')
            timeCounter += 1
            if (s[0]):
                data1.append(float(s[0]))
            if (s[1]):
                data2.append(float(s[1]))
```

```python
fig, axarr = plt.subplots(2, sharex=True)
plt.xlim(0,max(len(data1),len(data2)))
axarr[0].plot(data1)
axarr[0].set_title("Raw Data")
axarr[1].plot(data2)
avg1 = np.mean(data1)
axarr[0].set_ylim(avg1-1, avg1+1)
avg2 = np.mean(data2)
axarr[1].set_ylim(avg2-OFFSET, avg2+OFFSET)
fig.set_dpi(1000)
fig.set_size_inches(20,10)
```

```python
f_s = 1000 # Hz
fft_ar=[]

fft_x  = np.fft.fft(data1)
fft_ar = fft_x[50:]
freq   = np.fft.fftfreq(len(fft_x), f_s)

fft_x_shifted = np.fft.fftshift(fft_x)
freq_shifted = np.fft.fftshift(freq)
```

```python
fig, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(np.abs(fft_ar))
axarr[0].set_title("Frequency (Hz)")
axarr[1].plot(freq_shifted, fft_x_shifted)
fig.set_dpi(150)
fig.set_size_inches(20,10)
```

```python
t = scipy.linspace(0,1,1000)

FFT1 = [20 * scipy.log10(i) for i in abs(scipy.fft(data1))]
FFT2 = [20 * scipy.log10(i) for i in abs(scipy.fft(data2))]
freqs1 = scipy.fftpack.fftfreq(len(data1), t[1]-t[0])
freqs2 = scipy.fftpack.fftfreq(len(data2), t[1]-t[0])
```

```python
fig, axarr = plt.subplots(2, sharex=True, sharey=True)
axarr[0].plot(freqs1, FFT1, 'xb')
axarr[0].set_title("Frequency (Hz)")
axarr[1].plot(freqs2, FFT2, 'xr')

fig.set_dpi(150)
fig.set_size_inches(15,10)
```

```python
from scipy.signal import lfilter, firwin

sample_rate = 1000.
nyq_rate    = sample_rate / 2
cutoff_hz   = 1
numtaps     = 29

fir_coeff = firwin(numtaps, cutoff_hz / nyq_rate)
 
filtered_data1 = lfilter(fir_coeff, 1.0, data1)
filtered_data2 = lfilter(fir_coeff, 1.0, data2)
```

```python
fig, axarr = plt.subplots(2, sharex=True)
plt.xlim(0,max(len(filtered_data1), len(filtered_data2)))
axarr[0].plot(filtered_data1)
axarr[0].set_title("Filtered")
axarr[1].plot(filtered_data2)
avg1 = np.mean(filtered_data1)
axarr[0].set_ylim(avg1-OFFSET, avg1+OFFSET)
avg2 = np.mean(filtered_data2)
axarr[1].set_ylim(avg2-OFFSET, avg2+OFFSET)
fig.set_dpi(150)
fig.set_size_inches(15,10)
```

```python
def movingaverage(interval, window_size):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')

def raise_event(v1,window_size,th):
    events = [0] * len(v1)
    moving_average_v1 = movingaverage(v1,window_size)
    for i in range(len(v1)):
        if (abs(v1[i]-moving_average_v1[i])/abs(moving_average_v1[i]))>th :
            events[i] = -0.1
    return events

filtered_data1_av = movingaverage(filtered_data1, 1000)
filtered_data2_av = movingaverage(filtered_data2, 1000)
```

```python
fig, axarr = plt.subplots(2, sharex=True)
plt.xlim(0,max(len(filtered_data1_av),len(filtered_data2_av)))
axarr[0].plot(filtered_data1_av, 'r')
axarr[0].set_title("Filtered (Averages)")
axarr[1].plot(filtered_data2_av, 'r')
avg1 = np.mean(filtered_data1_av)
axarr[0].set_ylim(avg1-OFFSET, avg1+OFFSET)
avg2 = np.mean(filtered_data2_av)
axarr[1].set_ylim(avg2-OFFSET, avg2+OFFSET)
fig.set_dpi(150)
fig.set_size_inches(15,10)
```
